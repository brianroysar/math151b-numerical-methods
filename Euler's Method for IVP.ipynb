# Homework 1: Euler's Method for IVP

Given the following IVP: $y'= 2 + 2(2t-y)^2, 1 \leq t \leq \frac{3}{2}, y(1)=1$ , we can use the following to solve this IVP with varying step size $h$

```python
import numpy as np

### Set up ###

# f: ODE, y0: initial condition, h: step size
f = lambda t, y : (-)
y0 = 2
h = 0.25

# domain interval
t0 = 1
T = 2
interval = T-t0

# number of mesh points (N+1)
N = interval/h
mesh = int(N+1)

# generating t
t = np.linspace(t0,T, mesh)

# y
y = np.zeros((mesh))
y[0] = y0
### Eulers Method ###
for i in range(0, len(t)-1):
    y[i+1] = y[i] + h*f(t[i],y[i])

print(y)
```

```python
import matplotlib.pyplot as plt

print(y)

# Analytical Solution
y_analytical = lambda t: (2*t) + (1/(1-(2*t)))
y_act = y_analytical(t)

plt.plot(t, y, label="Euler's Method")
plt.plot(t, y_act, label = "Analytical Solution")
plt.legend()
```

Given another IVP, $y' = -5y + 3tsin(y), 0 \leq t \leq 1, y(0) = 3$ we can do the same thing.

```python
### Set up ###

# f2: ODE, y0: initial condition, h: step size
f2 = lambda t, y : -5*y + 3*t*np.sin(y)
y0 = 3
h = 0.25

# domain interval
t0 = 0
T = 1
interval = T-t0

# number of mesh points (N+1)
N = interval/h
mesh = int(N+1)

# generating t
t = np.linspace(t0,T, mesh)

# y
y = np.zeros((mesh))
y[0] = y0

### Eulers Method ###
for i in range(0, len(t)-1):
    print(t[i])
    y[i+1] = y[i] + h*f2(t[i],y[i])

print(y)
```
